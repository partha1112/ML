```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
dataset = pd.read_csv(r'C:\Users\Dell\Downloads\diabetes.csv')
```

```python
dataset.head()
```

```python
dataset.shape
```

```python
dataset['Outcome'].value_counts()
```

```python
dataset.groupby('Outcome').mean()
```

```python
# Separating data and labels
X = dataset.drop(columns = ['Outcome'], axis=1)
Y = dataset['Outcome']
```

```python
# Data Standardisation to make all the data bound to the same range so that our model perform better

scalar = StandardScaler()
```

```python
scalar.fit(X)
```

```python
X_tr = scalar.transform(X)
```

```python
# Train Test Split

x_train,x_test,y_train,y_test = train_test_split(X_tr,Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
# Training the model
```

```python
classifier = svm.SVC(kernel='linear')
```

```python
classifier.fit(x_train, y_train)
```

```python
# Model Evaluation
```

```python
x_train_prediction = classifier.predict(x_train)
x_train_accuracy = accuracy_score(x_train_prediction, y_train)
```

```python
x_train_accuracy
```

```python
x_test_prediction = classifier.predict(x_test)
x_test_accuracy = accuracy_score(x_test_prediction,y_test)
```
